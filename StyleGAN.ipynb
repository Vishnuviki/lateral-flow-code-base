```python
import cv2
import os
import numpy as np
import pandas as pd
from PIL import Image
import tensorflow as tf
```

```python
print(np.__version__)
print(tf.__version__)
```

```python
custom_dataset = os.path.join('/content/drive/My Drive/ARU/MajorProject')
resized_dataset = os.path.join('/content/drive/My Drive/ARU/styleGAN/resized-dataset/')
tf_dataset = os.path.join('/content/drive/My Drive/ARU/styleGAN/tf-records-dataset/')
```

```python
neg_images = os.listdir(custom_dataset);
post_images = os.listdir(custom_dataset + '/Positive');
neg_images_one = os.listdir(custom_dataset + '/Negative');
```

```python
WIDTH = 512
HEIGHT = 512
```

```python
neg_df_one = { 'id': neg_images_one}
neg_df_one = pd.DataFrame(neg_df_one)
```

```python
neg_ids_one = neg_df_one['id']
```

```python
NUM_ITEMS_NEG_ONE = len(neg_df_one)
NUM_ITEMS_NEG_ONE
```

```python
# Building data list
data_neg_one = []
for i in range(NUM_ITEMS_NEG_ONE):
    image = Image.open(custom_dataset + '/Negative/' + neg_ids_one[i])
    image = np.array(image.resize((WIDTH, HEIGHT)))
    data_neg_one.append(image)
    cv2.imwrite(resized_dataset + str(neg_ids_one[i]), image)
```

```python
neg_df = { 'id': neg_images}
neg_df = pd.DataFrame(neg_df)
neg_df = neg_df[neg_df.id != 'Positive'];
neg_df = neg_df[neg_df.id != 'Negative'];
```

```python
neg_ids = neg_df['id']
```

```python
NUM_ITEMS_NEG = len(neg_df)
NUM_ITEMS_NEG
```

```python
# Building data list
data_neg_one = []
for i in range(NUM_ITEMS_NEG):
    image = Image.open(custom_dataset + '/' + neg_ids[i])
    image = np.array(image.resize((WIDTH, HEIGHT)))
    data_neg_one.append(image)
    cv2.imwrite(resized_dataset + str(neg_ids[i]), image)
```

```python
post_df = { 'id': post_images}
post_df = pd.DataFrame(post_df)
post_ids = post_df['id']
```

```python
NUM_ITEMS_POST = len(post_df)
NUM_ITEMS_POST
```

```python
# Building data list
post_data = []
for i in range(NUM_ITEMS_POST):
    image = Image.open(custom_dataset + '/Positive/' + post_ids[i])
    image = np.array(image.resize((WIDTH, HEIGHT)))
    post_data.append(image)
    cv2.imwrite(resized_dataset + str(post_ids[i]), image)
```
